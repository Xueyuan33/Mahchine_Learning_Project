```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/Asurion_data_iPhone_v3.csv', parse_dates=['weeks_monday','month_year'])
```

```python
df
```

```python
newer_iphone_df = df[(df['phone_generation'] == 'iphone x') | (df['phone_generation'] == 'iphone 11') | (df['phone_generation'] == 'iphone 12') | (df['phone_generation'] == 'iphone 13') | (df['phone_generation'] == 'iphone 14')]
```

```python
release_dates = {
    'iphone x': '2017-11-03',
    'iphone 11': '2019-09-20',
    'iphone 12': '2020-10-23',
    'iphone 13': '2021-10-24',
    'iphone 14': '2022-09-16'
}
newer_iphone_df['days_since_release_date'] = newer_iphone_df.apply(lambda x: (x['weeks_monday'] - pd.to_datetime(release_dates[x['phone_generation']])).days, axis=1)
```

```python
newer_iphone_df
```

```python
newer_iphone_model_df = newer_iphone_df[(newer_iphone_df['phone_model'] == 'apple iphone x') |(newer_iphone_df['phone_model'] == 'apple iphone 11') | (newer_iphone_df['phone_model'] == 'apple iphone 12') | (newer_iphone_df['phone_model'] == 'apple iphone 13') | (newer_iphone_df['phone_model'] == 'apple iphone 14')]
newer_iphone_model_df
```

```python
iphone_no_cat_df = newer_iphone_model_df[['phone_model', 'weeks_monday', 'month', 'year', 'month_year', 'days_since_release_date','claim']]
iphone_no_cat_df 
```

```python
iphone_no_cat_df  = iphone_no_cat_df.groupby(['phone_model', 'weeks_monday', 'month', 'year', 'month_year', 'days_since_release_date'])['claim'].sum().reset_index()
iphone_no_cat_df  = iphone_no_cat_df.sort_values(by='weeks_monday')
iphone_no_cat_df[(iphone_no_cat_df['phone_model'] == 'apple iphone 14')]
```

```python
iphone_no_cat_no_feb13_df = iphone_no_cat_df[iphone_no_cat_df['weeks_monday'] != '2023-02-13']
iphone_no_cat_no_feb13_df.tail(10)
```

```python
iphone_no_cat_no_feb13_no_jun_28 = iphone_no_cat_no_feb13_df[iphone_no_cat_no_feb13_df['weeks_monday']!= '2021-06-28']
```

```python
iphone_no_cat_no_feb13_no_jun_28.to_csv('/data/p_dsi/teams2023/team2/asurion_data/time_series_ML_features_final.csv', index=False, date_format = '%Y-%m-%d')
```
