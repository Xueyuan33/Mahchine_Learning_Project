```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/Asurion_data_iPhone_v3.csv', parse_dates=['weeks_monday','month_year'])
```

```python
df = df[(df['weeks_monday'] != '2023-02-13') & (df['weeks_monday'] != '2021-06-28')]
```

```python
# df.set_index('weeks_monday')
# df.index = pd.to_datetime(df.index)
```

```python
# df = df.sort_index(ascending=True)
# df
```

```python
df['phone_size'] = df['phone_size'].str.replace('gb', '').astype(int)
```

```python
df['phone_size'].unique()
```

```python
df
```

```python
df['phone_generation'] = df['phone_generation'].str.replace('iphone ', '')
```

```python
df['phone_generation'].unique()
```

```python
df['phone_generation'] = df['phone_generation'].replace('x', '10')
```

```python
df = df[df['phone_generation'] != 'se']
```

```python
df['phone_generation'].unique()
```

```python
df['phone_generation'] = df['phone_generation'].astype(int)
```

```python
df['weeks_monday'] = pd.to_datetime(df['weeks_monday'])

df['week_of_year'] = df['weeks_monday'].apply(lambda x: x.isocalendar()[1]).astype(int)
```

```python
df['quarter'] = np.ceil(df['month'] / 3).astype(int)
```

```python
df['phone_generation'].unique()
```

```python
release_dates = {
    3: '2008-07-11',
    4: '2010-06-24',
    5: '2012-09-21',
    6: '2014-09-09',
    7: '2016-09-16', 
    8: '2017-09-22',
    10: '2017-11-03',
    11: '2019-09-20',
    12: '2020-10-23',
    13: '2021-10-24',
    14: '2022-09-16'
}
df['days_since_release_date'] = df.apply(lambda x: (x['weeks_monday'] - pd.to_datetime(release_dates[x['phone_generation']])).days, axis=1)
```

```python
df['holiday'] = df['weeks_monday'].apply(lambda x: 1 if x in pd.to_datetime(['2021-06-28', '2021-11-22', '2021-12-20',
                                                                             '2022-12-27', '2022-04-11', '2022-05-30',
                                                                             '2022-07-04', '2022-11-21', '2022-12-19',
                                                                             '2022-12-26']) else 0)
```

```python
df = df[['phone_model', 'phone_generation', 'phone_size', 'phone_color', 'week_of_year','month', 'quarter', 'year', 'holiday', 'days_since_release_date','claim']]
```

```python
one_hot = pd.get_dummies(df['phone_color'])
df = pd.concat([df, one_hot], axis=1)
df
```

```python
one_hot = pd.get_dummies(df['phone_model'])
df = pd.concat([df, one_hot], axis=1)
df
```

```python
df = df.drop(['phone_model', 'phone_color'], axis=1)
```
