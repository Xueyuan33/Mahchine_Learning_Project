```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
data = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/Asurion_data_v2.csv', parse_dates=['weeks_monday'])
print (data)
```

```python
data.describe()
```

```python
#iphone 10 dataset and timeseries
iphone_10 = data[data['phone_model'].str.contains("apple iphone x")]
iphone_10.describe()
#mean claims = 47.4
#iPhone X released: November 3, 2017
```

```python
iphone_10.groupby(['phone_color','phone_size']).count()
```

```python
sum_claim_10=iphone_10.groupby(['weeks_monday'], as_index=False)['claim'].sum()
sum_claim_10
```

```python
#timeseries for iphne X
plt.plot(sum_claim_10['weeks_monday'],sum_claim_10['claim'])
plt.xlabel('Date')
plt.ylabel('Claim')
plt.title('Iphone X time series')
plt.xticks(rotation=45, ha='right')
plt.show()
```
